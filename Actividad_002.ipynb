# Análisis descriptivo

## Carga de los datos y creacion del dataframe

```python
import pandas as pd
```

```python
# Base de Datos
df = pd.read_csv('https://raw.githubusercontent.com/Sifrisky/business_analytics/main/data_tata.csv')
```

```python
# Visualizacion de los primeros 5 registros
df.head()
```

## Transformacion de los datos

```python
# Revision del tipo de datos
df.info()
```

```python
# Conversion del tipo de datos object a float
df.history_purchase = pd.to_numeric(df['history_purchase'], errors = 'coerce')
```

```python
# Revision del tipo de datos
df.info()
```

```python
# Conversion del texto a primera letra en mayuscula
df.zone = df.zone.apply(lambda x: x.capitalize())
df.channel = df.channel.apply(lambda x: x.capitalize())
df.offer = df.offer.apply(lambda x: x.capitalize())
df.services = df.services.apply(lambda x: x.capitalize())
df.social_utm = df.social_utm.apply(lambda x: x.capitalize())
df.genre = df.genre.apply(lambda x: x.capitalize())
```

# Regresión Lineal

```python
# Para la regresion lineal, las observaciones tienen que ser iguales, rellenamos los valores nulos
```

```python
df.zip_code.fillna(method="ffill", inplace=True)
```

```python
df.info()
```
