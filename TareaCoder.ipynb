```python
#Limpiar los registros 
df.social_utm = df.social_utm.apply(lambda x: x.lower() )
```

```python
from pandas._libs.tslibs import conversion
#agrupamos por red social y armamos lista por red  y conversion

df.groupby(['social_utm', 'conversion'], as_index= False).count()

```
